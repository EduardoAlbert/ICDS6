```python
import pandas as pd
```

```python
titanicDf = pd.read_csv('titanic.csv')
titanicDf
```

```python
titanicDf.drop('Name', axis=1, inplace=True)
titanicDf
```

```python
titanicDf.drop('PassengerId', axis=1, inplace=True)
titanicDf
```

```python
titanicDf.drop('Ticket', axis=1, inplace=True)
titanicDf
```

```python
titanicDf.drop('Cabin', axis=1, inplace=True)
titanicDf
```

```python
titanicDf.replace('male', 0, inplace=True)
titanicDf.replace('female', 1, inplace=True)
titanicDf
```

```python
titanicDf.replace('S', 0, inplace=True)
titanicDf.replace('C', 1, inplace=True)
titanicDf.replace('Q', 2, inplace=True)
```

```python
titanicDf
```

```python
titanicDf = titanicDf.reset_index(drop=True)
titanicDf
```

```python
print(titanicDf[titanicDf['Survived']==0].mean())
```

```python
print(titanicDf[titanicDf['Survived']==1].mean())
```

```python
titanicDf
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(2)
```

```python
X = titanicDf.values[:,1:8]
```

```python
titanicDf
```

# Normalizando os dados

```python
from sklearn.preprocessing import normalize
```

```python
X = normalize(X)
X
```

## Aplicando o K-Means

```python
kmeans.fit(X)
previsao = kmeans.labels_
previsao
```

## DistribuiÃ§Ã£o dos grupos

```python
previsaoDf = pd.DataFrame(previsao)
print(previsaoDf[previsaoDf[0]==0])
```

```python
titanicDf['Kmeans'] = kmeans.labels_
titanicDf
```

# Entendendo melhor os grupos selecionados

```python
print(titanicDf[titanicDf['Kmeans']==0].mean())
```
